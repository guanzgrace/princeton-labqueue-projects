## remove_outliers 

Goal is to remove any request submitted before the shift starts, or after the shift ends. This should only be a problem with past data, as anything newer has the room password feature.

```python
label_to_starting_hour = dict()
label_to_starting_hour["isMonday"] = 19
label_to_starting_hour["isTuesday"] = 19
label_to_starting_hour["isWednesday"] = 19
label_to_starting_hour["isThursday"] = 19
label_to_starting_hour["isFriday"] = 19
label_to_starting_hour["isSaturday"] = 15
label_to_starting_hour["isSunday"] = 17
```

```python
def get_outliers():
    outlier_index = list()
    for k, v in label_to_starting_hour.items():
        outlier_index.append(df.where(df[k]==1).where(df["time_accepted"]==df["time_closed"]).where(df["time_created"].apply(get_hour) < v).dropna().index.values)
    return outlier_index
```

```python
outliers = get_outliers()
```

```python
drop_indexes = np.concatenate(outliers).ravel().tolist()
```

```python
df = df.drop(drop_indexes)
```

Additionally, we want to remove specific instances in the test data where the wait time of that individual was more than 2x every other person in that shift's wait time. This could correspond to students repeatedly putting themselves on the queue right after being helped. I manually searched for these pks.

```python
pks_to_remove = [18392, 19049, 18774, 19011, 18974, 19051, 18798, 19210, 19328]
df = df.where(~df["pk"].isin(pks_to_remove)).dropna()

df.to_csv("labqueue_anon_1.csv")
```
